```python
import pandas as pd
```

```python
df_customers = pd.read_csv('../../material/dados/olist_customers_dataset.csv')
df_customers.head(3)
```

```python
df_product = pd.read_csv('../../material/dados/olist_products_dataset.csv')
df_product.head(3)
```

```python
df_orders = pd.read_csv('../../material/dados/olist_orders_dataset.csv')
df_orders.head(3)
```

```python
df_items = pd.read_csv('../../material/dados/olist_order_items_dataset.csv')
df_items.head(3)
```

```python
df_full = pd.merge(df_customers, df_orders, on = 'customer_id', how = 'inner') 
df_full.info()
```

```python
df_completa = pd.merge(df_full, df_items, on = 'order_id', how = 'inner')
df_completa.info() 
```

```python
df_total = pd.merge(df_completa, df_product, on = 'product_id', how = 'inner')
df_total.info()
```

```python
df_total.to_csv('data.csv', index = False)
```
