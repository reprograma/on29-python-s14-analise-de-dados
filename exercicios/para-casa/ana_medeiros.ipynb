```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Bases do Olist utilizadas: Orders, Reviews e Payments

Relacionadas pelo 'order_id'

```python
# Fazendo a leitura do banco de dados de Pedidos

df_pedidos = pd.read_csv('olist_orders_dataset.csv')
df_pedidos
```

```python
# Fazendo a leitura do banco de dados de Pagamentos

df_pagamentos = pd.read_csv('olist_order_payments_dataset.csv')
df_pagamentos
```

```python
# Mesclando os bancos de dados de Pedidos e Pagamentos

df_olist = pd.merge(df_pedidos, df_pagamentos, on=['order_id'], how='inner')
df_olist
```

```python
# Fazendo a leitura do banco de dados de Resenhas

df_resenhas = pd.read_csv('olist_order_reviews_dataset.csv')
df_resenhas
```

```python
# Mesclando os 3 bancos de dados, dessa vez incluindo a de Resenhas para criar o banco de dados final

df_olist_final = pd.merge(df_olist, df_resenhas, on=['order_id'], how='inner')
df_olist_final
```

```python
df_olist_final.info
```

```python
columns_drop = ['order_purchase_timestamp',	'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date',	'payment_sequential', 'review_comment_title', 'review_comment_message',	'review_creation_date',	'review_answer_timestamp']
df_olist_final = df_olist_final.drop(columns_drop, axis=1)
df_olist_final
```

```python
# Fazendo a leitura da base final para anÃ¡lise

df_final = pd.read_csv('olist_final.csv')
df_final
```

Ranking do tipo de mÃ©todo de pagamento que costuma ser mais cancelado

```python
plt.figure(figsize=(12, 6))

sns.countplot(data = df_final[df_final['order_status'] == 'canceled'],
              x = 'payment_type',
              )

plt.title('Ranking do tipo de mÃ©todo de pagamento que costuma ser mais cancelado')
plt.xlabel('MÃ©todo de pagamento')
plt.ylabel('DistribuiÃ§Ã£o do mÃ©todo de pagamento mais cancelado')

plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.countplot(data = df_final[df_final['payment_type'] == 'credit_card'],
              x = 'payment_value',
              )

plt.title('Qual Ã© a avaliaÃ§Ã£o dos pedidos que foram entregues?')
plt.xlabel('AvaliaÃ§Ã£o')
plt.ylabel('DistribuiÃ§Ã£o das avaliaÃ§Ãµes de pedidos entregues')

plt.show()
```

Ranking de pedidos de avaliaÃ§Ã£o de pedidos que foram entregues

```python
plt.figure(figsize=(12, 6))

sns.countplot(data = df_final[df_final['order_status'] == 'delivered'],
              x = 'review_score',
              )

plt.title('Ranking de pedidos de avaliaÃ§Ã£o de pedidos que foram entregues')
plt.xlabel('AvaliaÃ§Ã£o')
plt.ylabel('DistribuiÃ§Ã£o das avaliaÃ§Ãµes de pedidos entregues')

plt.show()
```
