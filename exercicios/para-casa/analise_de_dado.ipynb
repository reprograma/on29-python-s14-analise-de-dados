```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_produtos = pd.read_csv('/content/olist_products_dataset.csv')
df_produtos
```

```python
df_produtos.info()
```

```python
columns_drop1 = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty','product_weight_g', 'product_length_cm', 'product_height_cm','product_width_cm']
df_produtos = df_produtos.drop(columns_drop1, axis=1)
df_produtos
```

```python
df_itens_pedidos = pd.read_csv('/content/olist_order_items_dataset.csv')
df_itens_pedidos
```

```python
df_itens_pedidos.info()
```

```python
columns_drop2 = ['order_id', 'order_item_id', 'shipping_limit_date']
df_itens_pedidos = df_itens_pedidos.drop(columns_drop2, axis=1)
df_itens_pedidos
```

```python
df_itens_pedidos["price"] = df_itens_pedidos["price"].replace({'\$': ''}, regex=True).astype(float)
```

```python
df_itens_pedidos["price"] = df_itens_pedidos["price"].apply(lambda x: f"${x:.2f}")
df_itens_pedidos
```

```python
df_itens_pedidos["freight_value"] = df_itens_pedidos["freight_value"].apply(lambda x: f"${x:.2f}")
df_itens_pedidos
```

```python
df_olist1 = pd.merge(df_produtos, df_itens_pedidos, on=['product_id'], how='inner')
df_olist1
```

```python
df_vendedores = pd.read_csv('/content/olist_sellers_dataset.csv')
df_vendedores
```

```python
df_vendedores.info()
```

```python
df_olist1 = pd.merge(df_olist1, df_vendedores, on=['seller_id'], how='inner')
df_olist1
```

```python
df_loc = pd.read_csv('/content/olist_geolocation_dataset.csv')
df_loc
```

```python
columns_drop3 = ['geolocation_lat', 'geolocation_lng']
df_loc = df_loc.drop(columns_drop3, axis=1)
df_loc
```

```python
df_loc = df_loc.rename(columns={"geolocation_zip_code_prefix": "seller_zip_code_prefix"})
df_loc
```

```python
df_olist1 = pd.merge(df_olist1, df_loc, on=['seller_zip_code_prefix'], how='inner')
df_olist1
```

```python
df_olist1.info()
```

```python
columns_drop4 = ['seller_city_x', 'seller_state_x','seller_zip_code_prefix','seller_city_y', 'seller_state_y']
df_olist1 = df_olist1.drop(columns_drop4, axis=1)
df_olist1
```

```python
df_olist1 = df_olist1.reset_index(drop=True)
df_olist1
```

Compreendo que não baixei a ordem de serviço logo não sei quais produtos de fato foram finalizados ou não mas, minhas analise não ira necessitar desse dados, pois são perguntas simples de facil repostas.

1- Quais foram os vendedores que mais intenção de venda e quero um contagem do decrescente

2-Qual estado teve mais intenção de compra

3-Qual categoria de produto tem mais procura

```python
categorias = df_olist1["product_category_name"].value_counts(normalize=True)
categorias
```

```python
categoria_porcento = categorias.sort_values(ascending=False)
categoria_porcento
```

Nessa analise vemos que a categoria mais vendida é cama mesa e banho , segundo lugar é beleza e saúde e em terceiro moveis e decoração

```python
plt.figure(figsize=(10, 6))
plt.bar(df_olist1["product_category_name"], color='skyblue')
plt.xlabel("Categoria de Produto")
plt.ylabel("Porcentagem")
plt.title("Distribuição de Vendas por Categoria de Produto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
