```python
import pandas as pd
```

```python
df_customers = pd.read_csv('../../material/dados/olist_customers_dataset.csv')
df_product = pd.read_csv('../../material/dados/olist_products_dataset.csv')
df_orders = pd.read_csv('../../material/dados/olist_orders_dataset.csv')
df_items = pd.read_csv('../../material/dados/olist_order_items_dataset.csv')
display(df_customers.head(2))
print(20*'**')
display(df_product.head(2))
print(20*'**')
display(df_orders.head(2))
print(20*'**')
display(df_items.head(2))
```

# Análises

## Compras feitas por cada estado

```python
#Estados com maior número de produtos comprados

df_compras_estado = df_total['customer_state'].value_counts().sort_values(ascending = False)
display(df_compras_estado.head(5))
```

```python
# Estados com menos número de produtos comprados
compras_estado = df_total['customer_state'].value_counts().sort_values(ascending = True)
display(compras_estado.head(5))
```

```python
import matplotlib.pyplot as plt

media_compras = compras_estado.mean() # Média de compras por estado

# Estados com número de compras acima e abaixo da média
estados_acima_media = compras_estado[compras_estado > media_compras]
estados_abaixo_media = compras_estado[compras_estado <= media_compras]

plt.figure(figsize=(10,6))
plt.bar(compras_estado.index, compras_estado.values, color='gray', label='Todos os Estados')
plt.bar(estados_acima_media.index, estados_acima_media.values, color='blue', label='Estados Acima da Média')
plt.bar(estados_abaixo_media.index, estados_abaixo_media.values, color='red', label='Estados Abaixo da Média')
plt.axhline(y=media_compras, color='green', linestyle='--', label=f'Média: {media_compras:.2f}')

plt.xlabel('Estado')
plt.ylabel('Número de Compras')
plt.title('Número de Compras por Estado - com base na média')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

## Proporção de categoria por produto

```python
proporcao_categorias = df_total['product_category_name'].value_counts(normalize=True).round(2)
print(proporcao_categorias)
```

```python
top_5_categorias = proporcao_categorias[:5]
outras_categorias = proporcao_categorias[5:].sum()
top_5_categorias['Outros'] = outras_categorias

plt.figure(figsize=(6, 4))
top_5_categorias.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='tab20', legend=False)
plt.ylabel('')
plt.title('Proporção das 5 Maiores Categorias de Produtos')
plt.axis('equal')
plt.show()
```

## Número de Compras e Valor Total das Vendas por Estado

```python
import seaborn as sns
dados_por_estado = df_total.groupby('customer_state').agg({
    'order_id': 'count',
    'price': 'sum'
}).rename(columns={'order_id': 'numero_compras', 'price': 'valor_total_vendas'}).reset_index()

plt.figure(figsize=(12, 6))
heatmap_data = dados_por_estado.pivot_table(index="customer_state", values="valor_total_vendas", aggfunc='sum')
heatmap_data = heatmap_data.sort_values(by="valor_total_vendas", ascending = False)
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=.5)
plt.xlabel('Estado')
plt.ylabel('Valor Total das Vendas (R$)')
plt.title('Mapa de Calor do Valor Total das Vendas por Estado')
plt.show()
```

**O estado de São Paulo  tem o maior valor total de vendas, com mais de 5,2 milhões de reais, seguido de Rio de Janeiro e Minas Gerais.**

**O mapa também mostra que as vendas estão mais concentradas nas regiões Sudeste e Sul do Brasil.**

        * As regiões Sudeste e Sul têm as maiores populações, a renda per capita mais alta e maior desenvolvimento econômico.





## Relação entre o frete e o valor do produto


```python
import matplotlib.pyplot as plt

plt.scatter(df_total['price'], df_total['freight_value'])
plt.xlabel('Preço do Produto')
plt.ylabel('Valor do Frete')
plt.title('Relação entre o Valor do Frete e o Preço do Produto')
plt.show()
```

        * A maioria dos produtos tem preços mais baixos e valor de frete mais baixos;
        * Há uma variação no valor do frete, sugerindo este não está diretamente relacionado ao preço do produto.

**Não parece haver uma relação linear forte entre o preço do produto e o valor do frete, pois há pontos dispersos e não seguem uma linha clara. Isso sugere que o valor do frete pode ser influenciado por múltiplos fatores além do preço do produto.**

## Tempo de entrega

```python
# Tempo médio de entrega 

tempo_medio_entrega = df_total['days_delivery_diff'].mean().round(0)
print(f"O tempo médio de entrega é de {tempo_medio_entrega} dias.\n")

# Total de pedidos com entrega atrasada

pedidos_atrasados = df_total[df_total['days_delivery_diff'] > 0].shape[0]
print(f"Foram entregues atrasados o total de {pedidos_atrasados} pedidos.")
```

**Embora a maioria das entregas seja realizada no prazo médio de 9 dias, deve-se ter um foco adicional em reduzir os atrasos para melhorar a experiência do cliente.**

## Status de entrega dos pedidos

```python
pedidos_por_status = df_total['order_status'].value_counts()
print(pedidos_por_status)
```

**A maioria esmagadora dos pedidos foi entregue com sucesso;
Isso sugere um bom desempenho geral na entrega, com algumas exceções que podem exigir atenção para melhorar a eficiência e reduzir cancelamentos.**

```python
pedidos_atrasados = df_total[df_total['days_delivery_diff'] > 0]['days_delivery_diff']

plt.figure(figsize=(6, 4))
plt.hist(pedidos_atrasados, bins=20, edgecolor='grey', alpha=0.9)
plt.xlabel('Dias de Atraso na Entrega')
plt.ylabel('Número de Pedidos')
plt.title('Distribuição dos Dias de Atraso na Entrega')
plt.grid(True)
plt.show()
```

        * A maioria dos pedidos tem um atraso de até 25 dias. 
        * À medida que o número de dias de atraso aumenta, o número de pedidos diminui . 
        * Há um pequeno número de pedidos que tem um atraso de mais de 175 dias.

**Pode-se observar essa proporção menor, mas não negligenciável, de pedidos com atrasos substanciais, sugerindo desafios em casos de logística ou processamento.**
