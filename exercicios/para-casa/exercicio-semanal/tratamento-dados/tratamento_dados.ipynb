```python
import pandas as pd
```

Quero 4 datasets:
- Pedidos;
- Itens dos pedidos;
- Dados do consumidor;
- Detalhes do pagamento dos consumidores.

```python
df_pedidos = pd.read_csv('olist_orders_dataset.csv')
df_pedidos
```

```python
df_itens = pd.read_csv('olist_order_items_dataset.csv')
df_itens
```

```python
df_consumidor = pd.read_csv('olist_customers_dataset.csv')
df_consumidor
```

```python
df_pagamento = pd.read_csv('olist_order_payments_dataset.csv')
df_pagamento
```

```python
#mesclando as tabelas pedidos e itens
df_olist_pedidos = pd.merge(df_pedidos, df_itens, on=['order_id'], how='inner')
df_olist_pedidos
```

```python
#mesclando olist_pedidos com a tabela de consumidores
df_olist_pedidos_uf = pd.merge(df_olist_pedidos, df_consumidor, on=['customer_id'], how='inner')
df_olist_pedidos_uf
```

```python
#mesclando as informações dos pedidos, itens e consumidores com as informações de pagamento
df_olist_pedidos_uf_pagamento = pd.merge(df_olist_pedidos_uf, df_pagamento, on='order_id', how='inner')
df_olist_pedidos_uf_pagamento
```

Explorando a tabela criada.

```python
df_olist_pedidos_uf_pagamento.info()
```

```python
df_olist_pedidos_uf_pagamento.isnull()
```

Filtrando a tabela

```python
colunas_deletadas = ['order_approved_at',
                     'order_estimated_delivery_date',
                     'customer_unique_id',
                     'product_id',
                     'seller_id',
                     'order_item_id',
                     'shipping_limit_date',
                     'customer_zip_code_prefix',
                     'payment_sequential',
                     'customer_id',
                     'price',
                     'order_purchase_timestamp',
                     'order_delivered_carrier_date',
                     'order_delivered_customer_date',
                     ]
```

```python
df_olist_pedidos_uf_pagamento = df_olist_pedidos_uf_pagamento.drop(colunas_deletadas, axis=1)
df_olist_pedidos_uf_pagamento
```

```python
df_olist_pedidos_uf_pagamento['order_status'].unique()
```

```python
df_olist_pedidos_uf_pagamento = df_olist_pedidos_uf_pagamento.loc[df_olist_pedidos_uf_pagamento['order_status'] == 'delivered']
df_olist_pedidos_uf_pagamento
```

```python
df_olist_pedidos_uf_pagamento.drop_duplicates(inplace=True)
df_olist_pedidos_uf_pagamento
```

```python
df_olist_pedidos_uf_pagamento = df_olist_pedidos_uf_pagamento.reset_index(drop=True)
df_olist_pedidos_uf_pagamento
```

```python
df_olist_pedidos_uf_pagamento.to_csv('olist_pedidos_uf_pagamento.csv', index=False)
```
