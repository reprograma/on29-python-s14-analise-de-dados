Para a nossa análise iremos considerar apenas os reviews que contém comentário. A intenção desse filtro é buscar as vendas, consumidores e principalmente as avalições mais reais possíveis.

Contudo, iremos checar primeiramente se a nossa "amostra" irá nos trazer parâmetros representativos dos nossos dados. Isto é, se a amostra escolhida baseada no critério de reviews com comentários, segue as medidas descritivas dos dados completos.

```python
df_avaliacoes_comentadas = df_avaliacoes.dropna(subset='review_comment_message')
df_avaliacoes_comentadas
```

```python
df_avaliacoes_comentadas.info()
```

Analisando os resultados encontrados, obtemos as seguintes conclusões:

* A média das pontuações das avaliações da amostra (3,670) é menor do que a média das pontuações no dados completos (4,086). Isso sugere que as avaliações que têm um comentário tendem a ser mais críticas.

* O desvio padrão da amostra é superior ao desvio padrão dos dados, indicando uma maior variabilidade das pontuações com comentário. 

* O intervalo interquartil é mais amplo na amostra (2,0 - 5,0) em comparação aos dados (4,0 - 5,0), sugerindo maior dispersão nas pontuações das avaliações comentadas. Esta é uma informação bastante importante para o embasamento do nosso filtro de avaliações comentadas, uma vez que aponta que o intervalo interquartil de todas as avalições gira somente entre a maiores notas (4-5), indicando um possível menor senso crítico destas avalições. Enquanto que, o intervalo interquartil mais amplo da amostra, com notas entre 2 e 5, sugere que estes compradores estão mais comprometidos em emitir uma opinião efetiva sobre os produtos. 

Apesar dessas discrepâncias não corroborarem para o critério de semelhança das medidas descritivas entre a amostra e os dados completos, elas apontam a maior criticidade dos usuários que deixam comentários em suas avaliações, contribuindo com a intenção inicial deste filtro aplicado.

Portanto, a amostra filtrada pode ser vista como representativa apenas de uma subpopulação específica — os clientes que deixam um comentário. Uma vez que ela não representa completamente a população geral de todas as avaliações, especialmente porque a média das pontuações é diferente. Esta subpopulação parece ser mais crítica e tem uma variabilidade maior nas avaliações. 

Contudo, para o interesse de embasar a nossa análise nas avaliações dos clientes que são mais engajados (aqueles que deixam comentários), esta amostra filtrada pode fornecer uma visão útil. Sendo assim, seguiremos a nossa análise com base na amostra escolhida.
