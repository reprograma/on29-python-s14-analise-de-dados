```python
import pandas as pd
```

```python
df_pagamentos_de_pedidos = pd.read_csv('olist_order_payments_dataset.csv')
df_pagamentos_de_pedidos
```

```python
df_pedidos = pd.read_csv('olist_orders_dataset.csv')
df_pedidos 
```

```python
df_detalhes_do_consumidor = pd.read_csv('olist_customers_dataset.csv')
df_detalhes_do_consumidor
```

```python
# Primeira junção entre df_pagamentos_de_pedidos e df_pedidos
df_merge1 = pd.merge(df_pagamentos_de_pedidos, df_pedidos, on='order_id', how='inner')
df_merge1
```

```python
# Segunda junção com df_detalhes_do_consumidor
df_merge2 = pd.merge(df_merge1, df_detalhes_do_consumidor, on='customer_id', how='inner')
df_merge2
```

```python
df_merge2.info()
```

```python
# convertendo o tipo de dados para data

df_merge2['order_delivered_carrier_date'] = pd.to_datetime(df_merge2['order_delivered_carrier_date'])
df_merge2['order_delivered_customer_date'] = pd.to_datetime(df_merge2['order_estimated_delivery_date'])
df_merge2['order_estimated_delivery_date'] = pd.to_datetime(df_merge2['order_estimated_delivery_date'])

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_merge2
```

```python
#  Contagem de forma de pagamento
forma_pagamento_counts = df_merge2['payment_type'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=forma_pagamento_counts.index, y=forma_pagamento_counts.values)
plt.title('Contagem de Forma de Pagamento')
plt.xlabel('Forma de Pagamento')
plt.ylabel('Número de Transações')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Contagem de clientes por cidade
cidade_counts = df_merge2['customer_city'].value_counts().head(10)  # Mostrando as top 10 cidades

plt.figure(figsize=(12, 6))
sns.barplot(x=cidade_counts.index, y=cidade_counts.values, palette='viridis')
plt.title('Número de Clientes por Cidade (Top 10)')
plt.xlabel('Cidade')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Contagem de clientes por estado
estado_counts = df_merge2['customer_state'].value_counts()

# Configurações do gráfico
plt.figure(figsize=(10, 6))
sns.barplot(x=estado_counts.index, y=estado_counts.values, palette='muted')
plt.title('Número de Clientes por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
