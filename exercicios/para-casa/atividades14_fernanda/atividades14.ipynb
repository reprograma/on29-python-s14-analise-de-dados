```python
import pandas as pd
```

```python
#Leitura do arquivo de pagamentos

df_pagamentos = pd.read_csv('../../dados brutos/olist_order_payments_dataset.csv')
df_pagamentos
```

```python
#Registros Nan?
df_pagamentos.info()
#Registros completos
```

```python
#Quais são as opções de tipo de pagamento?
df_pagamentos['payment_type'].unique()
```

```python
#Remover coluna de pagamento em sequência e do número de parcelas
remover_colunas = ['payment_sequential', 'payment_installments']
df_pagamentos = df_pagamentos.drop(remover_colunas, axis= 1)
```

```python
#Printar df limpo
df_pagamentos
```

```python
#Leitura do arquivo de pedidos
df_pedidos = pd.read_csv('../../dados brutos/olist_orders_dataset.csv')
df_pedidos
```

```python
#Registros Nan?
df_pedidos.info()
```

```python
#Registros incompletos, deletar linhas com NaN
df_pedidos = df_pedidos.dropna(subset=['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date'])
```

```python
#Resetar index
df_pedidos = df_pedidos.reset_index(drop=True)
df_pedidos
```

```python
df_pedidos.info()
#registros todos completos
```

```python
#Deixar info apenas dos pedidos cancelados
df_pedidos_cancelados = df_pedidos.loc[df_pedidos['order_status'] == 'canceled']
df_pedidos_cancelados = df_pedidos_cancelados.reset_index(drop=True)
df_pedidos_cancelados
```

```python
#Remover coluna de timestamp, carrier_date e status do pedido
remover_pgto = ['order_purchase_timestamp', 'order_delivered_carrier_date', 'order_delivered_customer_date']
df_pedidos_cancelados = df_pedidos_cancelados.drop(remover_pgto, axis= 1)
```

```python
#df com pedidos cancelados
df_pedidos_cancelados
```

```python
#Deixar info apenas dos pedidos entregues
df_pedidos = df_pedidos.loc[df_pedidos['order_status'] == 'delivered']
df_pedidos
```

```python
#Remover colunas
remover_pgto = ['order_status', 'order_purchase_timestamp', 'order_delivered_carrier_date', 'order_delivered_customer_date']
df_pedidos = df_pedidos.drop(remover_pgto, axis= 1)
```

```python
#Printar df limpo
df_pedidos = df_pedidos.reset_index(drop=True)
df_pedidos
```

```python
#Conectar tabelas de pagamentos e pedidos
df_pedido_pgto = pd.merge(df_pedidos, df_pagamentos, on= 'order_id', how= 'inner')
df_pedido_pgto
```

```python
#Leitura do arquivo de clientes
df_clientes = pd.read_csv('../../dados brutos/olist_customers_dataset.csv')
df_clientes
```

```python
#Conferir dados NaN
df_clientes.info()
```

```python
#Remover colunas
remover_info = ['customer_unique_id', 'customer_zip_code_prefix']
df_clientes = df_clientes.drop(remover_info, axis= 1)
```

```python
df_clientes
```

```python
#Juntar tabela de clientes
df_olist = pd.merge(df_pedido_pgto, df_clientes, on= 'customer_id', how= 'inner')
df_olist
```

```python
df_olist.info()
```

```python
#converter object em datetime nas colunas order_approved_at e order_estimated_delivery_date
df_olist['order_approved_at'] = pd.to_datetime(df_olist['order_approved_at'])
df_olist['order_estimated_delivery_date'] = pd.to_datetime(df_olist['order_estimated_delivery_date'])
df_olist
```

```python
#Converter data para ano/mês/dia (tirar horário)
df_olist['order_approved_at'] = df_olist['order_approved_at'].dt.strftime('%Y-%m-%d')
df_olist
```

```python
#Criar arquivo .csv com tabela que será usada nas análises
df_olist.to_csv('olist_final_dataset.csv', index=False)
```

**Perguntas a serem respondidas:**
1) Qual é o tipo de pagamento mais usado pelos clientes? 

2) Qual foi o mês de maior valor de compras em Osasco/SP? É possível estimar o motivo?

```python
#1) Tipo de pagamento mais usado pelos clientes:
df_olist['payment_type'].value_counts(normalize=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Gráfico das formas de pagamento

plt.figure(figsize=(12, 6))

sns.countplot(data = df_olist,
              x = 'payment_type')

plt.title('Formas de pagamento na Olist')
plt.xlabel('Tipo de pagamento')
plt.ylabel('Quantidade')

plt.xticks(rotation=330)
plt.show()
```

```python
# Criar coluna 'region' e agrupar os pedidos para depois plotar no gráfico
# as infos relacionando os tipos de pagamento por Estado.

df_olist.loc[df_olist['customer_state'].isin(['SP', 'ES', 'RJ', 'MG']), 'region'] = 'sudeste'
df_olist.loc[df_olist['customer_state'].isin(['PR', 'SC', 'RS']), 'region'] = 'sul'
df_olist.loc[df_olist['customer_state'].isin(['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO']), 'region'] = 'norte'
df_olist.loc[df_olist['customer_state'].isin(['MA', 'PI', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE', 'BA']), 'region'] = 'nordeste'
df_olist.loc[df_olist['customer_state'].isin(['GO', 'MT', 'MS', 'DF']), 'region'] = 'centro_oeste'

df_olist
```

```python
plt.figure(figsize=(12, 6))

sns.countplot(data = df_olist,
              x = 'payment_type',
              hue= 'region')

plt.title('Formas de pagamento mais utilizados por região')
plt.xlabel('Tipo de pagamento')
plt.ylabel('Quantidade')

plt.xticks(rotation=330)
plt.show()
```

```python
#Como os dados são majoritariamente sobre a região Sudeste, abaixo temos o gráfico considerando as demais regiões
# a fim de destacar as outras infos. 

df_br_parcial = df_olist.loc[df_olist['region'] != 'sudeste']

plt.figure(figsize=(12, 6))

sns.countplot(data = df_br_parcial,
              x = 'payment_type',
              hue= 'region')

plt.title('Formas de pagamento mais utilizados por região - exceto Sudeste')
plt.xlabel('Tipo de pagamento')
plt.ylabel('Quantidade')

plt.xticks(rotation=330)
plt.show()
```

```python
# Criar um df apenas para Osasco/SP

df_osasco = df_olist.loc[df_olist['customer_city'] == 'osasco']
df_osasco
```

```python
# Agrupar as compras por mês/ano para ver qual mês teve a maior soma
df_osasco.loc[:, 'order_approved_at'] = pd.to_datetime(df_osasco['order_approved_at'])

df_osasco['ano_mes'] = df_osasco['order_approved_at'].dt.to_period('M').astype(str)
df_soma = df_osasco.groupby('ano_mes').agg({'payment_value': 'sum'}).reset_index()

df_soma
```

```python
sns.set_theme(style='whitegrid')

plt.figure(figsize=(14, 7))
sns.lineplot(data=df_soma, x='ano_mes', y='payment_value', marker='o', color='blue')

plt.title('Total de Preço Gasto por Mês e Ano')
plt.xlabel('Ano e Mês')
plt.ylabel('Total de Preço Gasto')

plt.xticks(rotation=45)

plt.show()
```

**CONCLUSÕES**

1) Qual é o tipo de pagamento mais usado pelos clientes? 
O cartão de crédito é o meio de pagamento mais usado pelos clientes, cerca de 74% das compras. Em seguida, temos o boleto (aprox. 19%), voucher (aprox. 5,5%) e cartão de débito (aprox. 1,5%). Esse tipo de comportamento é encontrado em todas as regiões do Brasil, desde a região com mais dados (Sudeste) até a com menos dados (Centro-Oeste).

2) Qual foi o mês de maior valor de compras em Osasco/SP? É possível estimar o motivo?
O mês de maior valor de pagamento foi Maio/2018, com uma soma de R$9968,15. O fato pode ser justificado principalmente por ser comemorado o dia das Mães nesse mês. No gráfico, podemos observar que o segundo mês com maior valor foi Novembro/2017, que também pode ser justificado tanto pela Black Friday, quanto pelo Dia do Consumidor.
